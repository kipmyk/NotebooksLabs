```python
import numpy as np
import matplotlib.pyplot as plt

# Logistic map function
def logistic_map(x, r):
    return r * x * (1 - x)

# --- Part 1: Simulating the logistic map over time for a single r value ---

# Parameters
r = 3.9           # Growth rate
x0 = 0.5          # Initial population value
n = 100           # Number of time steps

# Generate data
x = np.zeros(n)
x[0] = x0
for i in range(1, n):
    x[i] = logistic_map(x[i - 1], r)

# Plot population over time
plt.figure(figsize=(10, 5))
plt.plot(range(n), x, marker='o', linestyle='-', markersize=2)
plt.title(f'Logistic Map Population Evolution (r = {r})')
plt.xlabel('Time Step')
plt.ylabel('Population (x)')
plt.grid(True)
plt.tight_layout()
plt.show()

# --- Part 2: Bifurcation diagram ---

# Range of r values
r_values = np.linspace(2.5, 4.0, 10000)
iterations = 1000   # Total iterations per r
last = 100          # How many points to keep (for stability visualization)

# Initialize x values
x = 1e-5 * np.ones_like(r_values)

# Prepare the plot
plt.figure(figsize=(12, 7))

# Iterate and collect last values
for i in range(iterations):
    x = logistic_map(x, r_values)
    if i >= (iterations - last):
        plt.plot(r_values, x, ',k', alpha=0.25)

# Finalize plot
plt.title("Bifurcation Diagram of the Logistic Map")
plt.xlabel("r (Growth Rate)")
plt.ylabel("x (Population)")
plt.grid(True)
plt.tight_layout()
plt.show()
```
