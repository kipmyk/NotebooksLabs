```python
import speedtest
import matplotlib.pyplot as plt

def Speed_Test():
    test = speedtest.Speedtest()

    download = test.download() / 10**6  # Mbps
    upload = test.upload() / 10**6      # Mbps
    ping = test.results.ping            # ms

    download = round(download, 2)
    upload = round(upload, 2)
    ping = round(ping, 2)

    print("Download Speed in Mbps:", download)
    print("Upload Speed in Mbps:", upload)
    print("Ping in ms:", ping)

    # Plotting
    labels = ['Download', 'Upload', 'Ping']
    values = [download, upload, ping]
    colors = ['skyblue', 'lightgreen', 'lightcoral']

    plt.figure(figsize=(6, 4))
    bars = plt.bar(labels, values, color=colors)
    plt.title('Internet Speed Test Results')
    plt.ylabel('Mbps / ms')

    # Add value labels above bars
    for bar in bars:
        height = bar.get_height()
        plt.text(bar.get_x() + bar.get_width()/2.0, height, f'{height}', ha='center', va='bottom')

    plt.tight_layout()
    plt.show()

Speed_Test()
```
